```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('ipldataset/matches.csv')
```

```python
ipl_df=pd.DataFrame(df)
```

## Team Performance

```python
team_wins=ipl_df['winner'].value_counts()
print("\nTeam Wins:\n",team_wins)
```

```python
# Runs scored analysis
average_runs_scored_per_match=ipl_df['target_runs'].mean()
print("\n Average Runs Scored per Match:",average_runs_scored_per_match)
```

```python
team_toss=ipl_df['toss_winner'].value_counts()
print("\nMost toss winners team:\n",team_toss)
```

```python
# Bar Chart of Team wins
plt.Figure(figsize=(10,13))
sns.countplot(y='winner',data=ipl_df)
plt.title('Number of Wins by a team')
plt.xlabel("Number Of Wins")
plt.ylabel("Teams")
plt.show()
```

```python
# Histogram of runs scored
plt.Figure(figsize=(8,5))
plt.hist(ipl_df['target_runs'],bins=9)
plt.title('Distribution of runs scored per match')
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.show()
```
